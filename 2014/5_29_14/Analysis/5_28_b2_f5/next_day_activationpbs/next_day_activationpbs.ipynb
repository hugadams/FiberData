### DIRECTIONS:

**In the cell below, toggle <font color='blue'>INTERACTIVE</font> to switch between matplotlib and plotly plots.  Choose your desired <font color='blue'>INTERVAL</font> unit, and I'd recommend setting  <font color='red'>SAMPLE_BY</font> if the dataset is over 200 curves.  In <font color='blue'>SLICE THE DATA</font>, set time slices and a simple description of each data subset, then run the cells below.** 

```python
INTERACTIVE = True 
INTERVAL = 'm' #hr, m, s, d
SAMPLE_BY = 1 #2,4,10 etc...

def six_plot_inter(*args, **kwargs):
    """ Interactive six_plot_inter """
    f = six_plot(*args, **kwargs)
    if INTERACTIVE:
        fname = op.join("~/Desktop/Dropbox/FiberData/2014/5_29_14/Datafiles", "next_day_activationpbs", kwargs.pop('title', ''))
        fname=fname.replace('/', '_').lstrip('_')
        py.iplot_mpl(f, 
                     filename = fname,
                     overwrite=True,
                     resize=False, 
#                     world_readable=False,
                    height=650,
                     update=_pyupdate) #Overwrite some mpl-conversion settings
```

### FULL DATA

```python
ts = TimeSpectra.from_csv('next_day_activationpbs_cropped.csv', index_col=0)
ts = ts.ix[:, ::SAMPLE_BY]
ts.reference = 0
ts.specunit = 'nm'
ts = ts.as_interval(INTERVAL)

six_plot_inter(ts, title='Full Data %s-%s%s' % (ts.columns[0], ts.columns[-1], INTERVAL), striplegend=True)
```

### SLICE THE DATA

```python
SLICES = ( 
          (12.15,40, 'EDC prior to erratic'),  
          (40,50, 'Erratic EDC'), 
          (50, 58, 'Post EDC stability; no BSA BINDING') 
            )

INTERACTIVE = False #Want static plots from here down
```

**(Add longer description of SLICES here if desired)**

```python
for (start, end, desc) in SLICES:
    ts_slice = ts.iloc[:, start:end]
    ts_slice.reference = 0 #Should be automatic when slicing?
    six_plot_inter(ts_slice, title="Timeslice: %s-%s%s (%s)" % (start, end, INTERVAL, desc))
```
